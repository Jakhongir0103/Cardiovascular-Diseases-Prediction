```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
x_train=pd.read_csv('../../Master_courses/ML_course/projects/project1/data/dataset/x_train.csv')
y_train=pd.read_csv('../../Master_courses/ML_course/projects/project1/data/dataset/y_train.csv')
```

```python
x_train.head()
```

```python
health = ['GENHLTH','PHYSHLTH','MENTHLTH','POORHLTH']
health_care_access = ['PERSDOC2','MEDCOST','CHECKUP1','_HCVU651']
hypertension = ['_CHOLCHK','_RFCHOL']
```

## health

```python
x_health = x_train[health]
x_health.head()
```

```python
x_health.corr()
```

```python
x_health.loc[(x_health.GENHLTH==9) | (x_health.GENHLTH==7), 'GENHLTH'] = np.nan
x_health.fillna(x_health.GENHLTH.mean(),inplace=True)
```

```python
x_health.loc[(x_health.PHYSHLTH.isna()) | (x_health.PHYSHLTH==99) | (x_health.PHYSHLTH==88) | (x_health.PHYSHLTH==77), 'PHYSHLTH'] = 0
x_health.loc[(x_health.MENTHLTH.isna()) | (x_health.MENTHLTH==99) | (x_health.MENTHLTH==88) | (x_health.MENTHLTH==77), 'MENTHLTH'] = 0
x_health.loc[(x_health.POORHLTH.isna()) | (x_health.POORHLTH==99) | (x_health.POORHLTH==88) | (x_health.POORHLTH==77), 'POORHLTH'] = 0
```

```python
x_health.corr()
```

```python
x_health_care_access.loc[(x_health_care_access._HCVU651==9), '_HCVU651'] = 0
```
