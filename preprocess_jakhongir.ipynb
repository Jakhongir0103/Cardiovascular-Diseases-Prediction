```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)

import sys
sys.path.insert(0, './util')

import preprocess as pp
```

```python
x_train=pd.read_csv('../course/projects/project1/data/dataset/x_train.csv',)
y_train=pd.read_csv('../course/projects/project1/data/dataset/y_train.csv')
```

```python
x_train.head()
```

```python
# health
health = ['GENHLTH','PHYSHLTH','MENTHLTH','POORHLTH']
print('before')
print(x_train[health].corr(),end='\n\n')
x_train['GENHLTH'] = pp.fill_null(x_train,'GENHLTH',[7,9])
x_train['PHYSHLTH'] = pp.fill_null(x_train,'PHYSHLTH',[77,88,99],method='zero')
x_train['MENTHLTH'] = pp.fill_null(x_train,'MENTHLTH',[77,88,99],method='zero')
x_train['POORHLTH'] = pp.fill_null(x_train,'POORHLTH',[77,88,99],method='zero')
print('after')
print(x_train[health].corr())
```

```python
# health care access
health_care_access = ['PERSDOC2','MEDCOST','CHECKUP1','_HCVU651']
print('before')
print(x_train[health_care_access].corr(),end='\n\n')
x_train['PERSDOC2'] = pp.fill_null(x_train,'PERSDOC2',[7,9])
x_train['MEDCOST'] = pp.fill_null(x_train,'MEDCOST',[7,9])
x_train['CHECKUP1'] = pp.rescale(x_train,'CHECKUP1',{8:5})
x_train['CHECKUP1'] = pp.fill_null(x_train,'CHECKUP1',[7,9])
x_train['_HCVU651'] = pp.fill_null(x_train,'_HCVU651',9,method='zero')
print('after')
print(x_train[health_care_access].corr())
```

```python
# hypertension
hypertension = ['_CHOLCHK','_RFCHOL']
print('before')
print(x_train[hypertension].corr(),end='\n\n')
x_train['_RFCHOL'] = pp.fill_null(x_train,'_RFCHOL',9,method='zero')
x_train['_CHOLCHK'] = pp.fill_null(x_train,'_CHOLCHK',9,method='zero')
print('after')
print(x_train[hypertension].corr())
```
